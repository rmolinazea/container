##Differences in Color vs Gray Scale

There are a few difference to note, so lets go thorugh them and explain what they mean.


*   Our gray image's shape only had 2 numbers, (3024, 4032). Our color image contains 3 numbers (3024, 4032, 3). The three dimensions represent the number of channels in our image. The most common format for colored images is RGB. 
*   The other difference is in the size of our images. Our colored images contains 3 times more pixels than the grayscale image, hence it grayscale images are often used to reduce computational complexity.  



##Changing Image Format

As we learned previously, colored images have a large computational complexity hence grayscale images are often prefered. We can convert our colored images into grayscale, using the *rgb2gray* method.

1. Firstly we will import the *rgb2gray* method from the skimage.color library. 
```
from skimage.color import rgb2gray 
``` 
2.   Next we use the *imread* method to import our image.
```
img = imread('cat.jpeg')
```
3. We now create a new variable to hold the converted image.
```
img_new = rgb2gray(img)
```
4. We use plt.subplot to show both images using the imshow method.
```
plt.subplot(121), imshow(img)
plt.title('RGB Format')
```
```
plt.subplot(121), imshow(new_img)
```

````python
from skimage.color import rgb2gray
img = imread('cat.jpeg')
img_new = rgb2gray(img)

plt.subplot(121), imshow(img)
plt.title('RGB Format') 

plt.subplot(122), imshow(img_new)
plt.title('Grayscale Format') 

plt.show()
````

Besides RGB and Grayscale, there are other image formats including HSV(Hue Saturation Value), and HSL (Hue Saturation Lightness) which are alternative representations of RGB.

- Hue is a degree on the color wheel where 0 is for red, 120 is green, 240 is blue and again 360 would be red.
- Saturation represents the percentage of that color, where 0 is white and 100  is the full color.
- Value denotes the mixture of the colors with varying amounts of black or white paint.
- Lightness is another way to show the shade of the image where 0 is black and 1 is white.

See the images provided here for further clarification:

##Resizing Images 

One of the [biggest challenges](https://www.analyticsvidhya.com/blog/2019/09/9-powerful-tricks-for-working-image-data-skimage-python/#:~:text=Scikit%2Dimage%2C%20or%20skimage%2C,easy%20to%20learn%20and%20use.) in computer vision is that we require a huge amount of data for training our model. The data we collect is often from different sources which might result in variation in the size of the images. This might be a problem while extracting features from the images, or using the same for data augmentation.

When we are training our models it is important to make sure all images are the same size, and we will reshize our images using skimage. Hence resizing images is one of the most important steps in preprocessing.

We are going to use the resize function from skimage. The input to this function will be the image we want to update and the required dimensions for the new image:


````python
from skimage.transform import resize
img = imread('cat.jpeg')
#resize image
img_resized = resize(img, (300, 300))

#plot images
plt.subplot(121), imshow(img)
plt.title('Original Image')
plt.subplot(122), imshow(img_resized)
plt.title('Resized Image')
plt.show()
````

## Rescaling(Upscaling/Downscaling) Images

This implies scaling the image by a particular factor, for example downscaling an image by half or upscaling an image by 2. 

##Why use rescaling when we can use resize?
If our images were all the same size, for example (400,400), we could easily use the resize function to make all the images (200,200) but if our images are not all the same size we cannot simple resize them because 'half' of each image would be different.

*   If all images are the same size, we can use resize.
*   If all images are different sizes, we use rescale.

````python
from skimage.transform import rescale
img = imread('cat.jpeg')
img_rescaled = rescale(img, scale=(0.5,0.5,1))
#The parameter is either one value, or one value per spatial axis. Here because we need the top axis to be the same so we can add the title we say 1.


plt.subplot(121), imshow(img)
plt.title('Original Image')

plt.subplot(122), imshow(img_rescaled)
plt.title('Rescaled Image')

plt.show()
````

## Rotating Images by Different Angles

Images may not be at the correct angle, so we use the rotate method to angle them correctly.



````python
from skimage.transform import rotate
image = imread('tilt_image.png')

image_rotated = rotate(image, angle=45)
imshow(image_rotated)
````

Although our orienation problem is fixed, we can see our image has been cropped. Here we are not losing any important information but that is not always the case. The issue is handeled by the resize parameter in the rotation method.

We can use the rotation method for data augmentation, which is a technique of generating more samples for a training model, using the available data. 

See [here](https://www.analyticsvidhya.com/blog/2019/09/9-powerful-tricks-for-working-image-data-skimage-python/#:~:text=We%20can%20also,classified%20as%20%E2%80%98cat%E2%80%99%3A) for an example and further explanation. 


````python
from skimage.transform import rotate
image = imread('tilt_image.png')

image_rotated = rotate(image, angle=45, resize=True)
imshow(image_rotated)
````

## Flip Images Horizontally and Vertically

We can flip an image both horizontally and vertically. This creates a mirror image along the horizontal/vertical axis. 

There is no direct function/method for this is skimage, but we can use NumPy. 

NumPy provides functions *flipud* and *fliplr* for flipping the images across the *horizontal* and *vertical* axis respectively.


````python
from numpy import fliplr, flipud
cat = imread('cat.jpeg')

cat_flip = fliplr(cat)

plt.subplot(141), imshow(cat)
plt.subplot(142), imshow(cat_flip)

plt.show()
````

## Cropping Images

We can crop our images using 2 different methods:

(The image in the example is (1067 x 1600)

*   First, simply specify the new image shape you want. For this example, it would be image [100:967, 100:1500]. This will work fine for a single image. What if we have multiple images? We would have to mention the new image shape for every image (not ideal).
*   Another way could be to calculate the cropped image dimension using the current image shape, which can be determined using the command image.shape. So the height of the new image would be [100:img.shape[0]-100] and the width will be [100:img.shape[1]-100].



````python
image = imread('cat.jpeg')

# selecting part of the image only 
cropped = image[100:(img.shape[0]-100),100:(img.shape[1]-100)]

plt.subplot(121), imshow(image)
plt.title('Original Image')
plt.subplot(122),imshow(cropped)
plt.title('Cropped Image')

plt.show()
````

##Altering the Brightness of Images

The brightness of images can be changed using the *adjust_gamma* function, which uses a method called **gamma correlation**. For any given image, the pixel values are first normalized between 0 – 1 and then multiplied by a specified gamma value. The resulting pixel values are scaled back to the range 0-255.

For gamma *greater than 1*, the output image will be *darker* than the input image. While for gamma *less than 1*, the output image will be *brighter* than the input image.




````python
from skimage import exposure

#adjusting brightness
image = imread('cat.jpeg')
image_bright = exposure.adjust_gamma(image, gamma=0.5,gain=1)
image_dark = exposure.adjust_gamma(image, gamma=1.5,gain=1)

# plotting images
plt.subplot(131), imshow(image)
plt.title('Original Image')

plt.subplot(132),imshow(image_bright)
plt.title('Bright Image')

plt.subplot(133),imshow(image_dark)
plt.title('Dark Image')

plt.show()
````

## Applying Fliters

We can use Filters to modify or enhance an image’s features. We can use filters for various purposes, such as smoothing and sharpening the image, removing noise, highlighting features and edges in the image, etc.

When a filter is applied to an image, every pixel value is replaced by a new value generated using surrounding pixel values. 

The simplest filter is a median filter, where the pixel values are replaced with the median of neighboring pixels.



````python
from skimage.filters import median

image = imread('cat.jpeg', as_gray=True)
image_median = median(image)

# plotting images
plt.subplot(121), imshow(image)
plt.title('Original Image')

plt.subplot(122),imshow(image_median)
plt.title('Smooth Image')

plt.show()
````

Another popular filter used for highlighting edges is the sobel filter, to learn how it works click [here](https://www.analyticsvidhya.com/blog/2019/08/3-techniques-extract-features-from-image-data-machine-learning-python/).



````python

from skimage.filters import sobel_h

image = imread('cat.jpeg', as_gray=True)
image_sobelh = sobel_h(image)

# plotting images
plt.subplot(121), imshow(image)
plt.title('Original Image')

plt.subplot(122),imshow(image_sobelh, cmap=plt.cm.gray)
plt.title('Horizontal Edge')

plt.show()
````
